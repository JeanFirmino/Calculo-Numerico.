```
# 1/3 de simpson
# função para integrar
import numpy as np
def f(x):
    return ((x-1)**2 + np.log(x))
# implementando a regra de 1/3 de simpson
def regra_de_simpson( limite_inferior, limite_superior, sub_intervalo):
    # Calculando o intervalo
    h = (limite_superior - limite_inferior) / sub_intervalo
    # encontrando a soma 
    integracao = f(limite_inferior) + f(limite_superior)
    for i in range(1,sub_intervalo):
        k = limite_inferior + i*h
        if i%2 == 0:
            integracao = integracao + 2 * f(k)
        else:
            integracao = integracao + 4 * f(k)
    # encontrando valor final da intergração
    integracao = integracao * h/3
    return integracao
resultado1 = 0
def f4(x):
    return (-6*(x**(-4)))
def errosimpa1(limite_inferior,limite_superior):
    suporte = []
    i = limite_inferior
    while i<=limite_superior:
        if i>=limite_inferior and i<=limite_superior:
            suporte.append(f4(i))
        i += 0.000001
    return (((-1/2880)*(max(suporte)))*((limite_superior - limite_inferior)**5))
j = 0
for sub_intervalo in range(1,100):
    # entradas da integração
    limite_inferior = 1
    limite_superior = 4
    # Chamando a funcao pela regra de 1/3 de simpson
    resultado1 = regra_de_simpson( limite_inferior, limite_superior, sub_intervalo)
    if sub_intervalo==1:
        print(f'x\t\t\tn\t\t\terro relativo\t\t\t\n')
    erro = (np.absolute(resultado0 - resultado1)/np.absolute(resultado0))
    print(f'{resultado1}\t\t{sub_intervalo}\t\t{erro}\n')
    if erro >= 10**(-4):
        continue
    elif erro < 10**(-4) and j==0:
        j = sub_intervalo
        continue
    else:
        continue
print(f'Menor número de subintervalos é {j}\n')
print(f'O erro de 1/3 de simpson possuí um erro de {errosimpa1(limite_inferior,limite_superior)}')
```
