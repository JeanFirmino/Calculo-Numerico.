```
# Método do Trapézio 2.a.5pontos
# Definir a função de integração
import numpy as np
def f(x):
    return np.exp(2*x)
# erro da regra do trapézio
def f2(x):
    return 4*(np.exp(2*x))
def errotrap1(limite_inferior,limite_superior):
    suporte = []
    i = limite_inferior - 1
    while i<=limite_superior:
        if i>=limite_inferior and i<=limite_superior:
            suporte.append(f2(i))
        i += 0.00001
    return (-(max(suporte)/12)*((limite_superior - limite_inferior)**3))
# Aplicando o método do trapézio
def metodo_do_trapezio1( limite_inferior, limite_superior, sub_intervalo):
    # dividindo em intervalos
    h = (limite_superior - limite_inferior) / sub_intervalo
    # encontrando a soma
    integracao = f(limite_inferior) + f(limite_superior)
    for i in range(1,sub_intervalo):
        k = limite_inferior + i*h
        integracao = integracao + 2 * f(k)
    # encontrando o valor da integração final
    integracao = integracao * h/2
    return integracao
# Entradas do intervalo e do sub intervalo
limite_inferior = 1/2
limite_superior = 4
sub_intervalo = 5
# Chamamento do metodo do trapézio
t1 = 0
t1 = metodo_do_trapezio1( limite_inferior, limite_superior, sub_intervalo)
print(f'Integração pelo método do trapézio é igual a {t1}\n')
t2 = errotrap1(limite_inferior,limite_superior)
print(f'O erro do trapézio é {t2} \n')
print(f'O erro relativo é {np.absolute(t0 - t1)/np.absolute(t0)}\n')
```
