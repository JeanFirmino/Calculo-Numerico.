```
# DIFERENÇAS DIVIDIDAS (Newton) 6.a)
# para encontrar o produto dos termos
from sympy import *
def produto(i, xn, x):
	prod = 1
	for j in range(i):
		prod = prod * (xn - x[j])
	return prod
# Para carcular a tabela das diferenças divididas
def calculandotab(x, y, n):
	for i in range(1, n):
		for j in range(n - i):
			y[j][i] = ((y[j][i - 1] - y[j + 1][i - 1]) / (x[j] - x[i + j]))
	return y
# aplicando newton, metodo da diferenças divididas
def formula(xn, x, y, n):
	soma = y[0][0]
	for i in range(1, n):
		soma = soma + (produto(i, xn, x) * y[0][i])
	return soma
# apresentando a tabela
def apresentatab(y, n):
	for i in range(n):
		for j in range(n - i):
			print(round(y[i][j], 4), "\t", end = " ")
		print("")
#Dados da questão
#n numero de pares x e y
n = 6
x = [ 1.3, 1.4, 1.5, 1.6, 1.7, 1.8 ]
y = [[0 for i in range(n)] #criando a matriz y de zeros 
        for j in range(n)]
#y[][0] usado para iniciar a tabela das diferenças divididas
y[0][0] = 0.9929
y[1][0] = 0.9252
y[2][0] = 0.7781
y[3][0] = 0.5494
y[4][0] = 0.2489
y[5][0] = -0.0982
# calculando a tabela
y=calculandotab(x, y, n);
# Apresentando a tabela
apresentatab(y, n)
# Valor para interpolar 
xn = Symbol('xn') # dado na questão
yn = formula(xn, x, y, n)
print(f'\n Interpolando pelo método de Newton\n Y = {simplify(yn)}\n')
```

```
# Interpolação de Newton-Gregory 6.b)
# Calculando V
def calculav(v, n): # Serve como auxiliar para a função de newton d1,d2,d3...
	temporizador = v
	for i in range(1, n):
		temporizador = temporizador*(v - i) 
	return temporizador
# Serve para calcular o fatorial necessario
def fatorial(n):
	f = 1
	for i in range(2, n + 1):
		f *= i
	return f
#Dados da questão
n = 6
x = [ 1.3, 1.4, 1.5, 1.6, 1.7, 1.8 ]
y = [[0 for i in range(n)] #criando a matriz y de zeros 
        for j in range(n)]
#y[][0] usado para iniciar a tabela das diferenças divididas
y[0][0] = 0.9929
y[1][0] = 0.9252
y[2][0] = 0.7781
y[3][0] = 0.5494
y[4][0] = 0.2489
y[5][0] = -0.0982
# Calculando a diferenças divididas
# A tabela
for i in range(1, n):  # n é a quantidade de pares da questão 
	for j in range(n - i): #i e j são apenas os indexadores
		y[j][i] = y[j + 1][i - 1] - y[j][i - 1]
# Apresentando a tabela
for i in range(n): 
	print(x[i], end = "\t") # valores de x  
	for j in range(n - i): 
		print(y[i][j], end = "\t") # valores calculados de y onde os primeiros valore, 
                                    #pelo metodo de newton são os d1,d2,d3...
	print("")
# Valor para interpolar
xng = Symbol('xng')
# iniciando o somatório de newton 
soma = y[0][0]
v = (xng - x[0]) / (x[1] - x[0])
for i in range(1,n): 
	soma = soma + (calculav(v, i) * y[0][i]) / fatorial(i)
yng = soma
print(f'\n Interpolação de Newton-Gregory \n Y = {simplify(yng)}\n')
```
