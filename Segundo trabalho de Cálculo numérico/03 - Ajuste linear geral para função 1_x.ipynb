```
import numpy as np
# V^T.a=V^t.y ajuste linear geral 
# X e Y anunciados pela questão
# xi = np.array([ 0.0, 2.5, 5.0, 7.5, 10.0, 12.5], dtype='double')
# yi = np.array([ 110, 85, 73, 65, 59, 47], dtype='double')
#3º
# FUNÇÃO DADA f(x)= a/(b+x)
'''
Temos que inverter todos os Y, pois para linearizar essa função
foi preciso inverter 1/x para que se transforma em x. mas se caso 
quiser inverter o y ao invés do x também funcionaria.
linearizando a função temos:
Y = (a/b) + a.(1/x)
1/Y = (b/a) + (1/a).X
'''
xi = np.array([ 0.0, 2.5, 5.0, 7.5, 10.0, 12.5], dtype='double')
#yi = np.array([ 110, 85, 73, 65, 59, 47], dtype='double')  INICIAL
yi = np.array([ 1/110, 1/85, 1/73, 1/65, 1/59, 1/47], dtype='double')  #INVERTIDO (1/Y)
V = np.array([ xi**0, xi], dtype='double').transpose()
#V^t.V.aa=V^tY
##aa=(V^t.V)^1.V^t.y
aa = (np.linalg.inv((V.transpose()).dot(V))).dot(V.transpose()).dot(yi)
o=['b/a','1/a']
print(f'\n\n X={xi}\n\n Y={yi}\n\n V={V}\n\n A={aa}\n\nA={o}\n\n')
print(f'''Para encontrar 'a' basta fazermos uma substituição cruzada em 1/a={aa[1]},\n 
o que resulta em a={1/aa[1]}\n
Agora para encontrar 'b' podemos substituir 'a' em b/a\n
Que será b={aa[0]*(1/aa[1])}\n
''')
```
