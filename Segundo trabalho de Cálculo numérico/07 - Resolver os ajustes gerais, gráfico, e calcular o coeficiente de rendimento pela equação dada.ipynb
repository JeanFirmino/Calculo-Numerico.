```
import numpy as np
# V^T.a=V^t.y ajuste linear geral 
# X e Y anunciados pela questão
xi = np.array([ -6, -4, -2, 0, 2, 4, 6], dtype='double')
yi = np.array([ 6, 1, -3, -1, 2, 9, 14], dtype='double')
#--------7.a)-----------------------------------------------------------------------------
# FUNÇÃO DADA PELA LETRA 7.a 
# u1(x) = a*x**2 + b*x <--- V1
V1 = np.array([ xi**2, xi**1], dtype='double').transpose()
##aa=(V^t.V)^1.V^t.y
aa = ((np.linalg.inv((V1.transpose()).dot(V1))).dot(V1.transpose())).dot(yi)
a = aa[0]
b = aa[1]
o=[ ' a', ' b']
print(f'\n\n X = {xi}\n\n Y = {yi}\n\n V1 = {V1}\n\n a = {aa}\n\n aa = {o}\n\n')
#---------------------------------------------------------------------------------------#
#----------7.a)
#função dada pela questão
# u2(x) = c*x**2 + d   <--- V2
V2 = np.array([ xi**2, xi**0], dtype='double').transpose()
##ab=(V^t.V)^1.V^t.y
ab = ((np.linalg.inv((V2.transpose()).dot(V2))).dot(V2.transpose())).dot(yi)
c = ab[0]
d = ab[1]
l=[ ' c', ' d']
print(f'\n\nV = {V2}\n\nab = {ab}\n\na = {l}\n\n')
## 8.b
#dado pela questão d= somatorio(Yi-U1(xi)) de=1 até len(yi)
somau1d=0
somau2d=0
for i in range(len(yi)):
    somau1d += ( yi[i] - ( a*(xi[i]**2) + b*xi[i] ) )**2
    somau2d += ( yi[i] - ( c*(xi[i]**2) + d ) )**2
print(f'O D calculado para U1(xi) = {somau1d}\n e para U2(xi) = {somau2d}\n\n')
#>>>>>>>>>>>>>>>>> COMO CRIAR GRAFICOS !!! <<<<<<<<<<<<#
x = np.arange(-6,6,0.00001)
u1 = a*(x**2) + b*x
u2 = c*(x**2) + d
import matplotlib.pyplot as plt #biblioteca de graficos do python
plt.figure(figsize=(12, 12)) #criar o quadro branco de tamanho 10x10
plt.style.use('classic') #o estilo de cores do grafico
plt.scatter(xi,yi,s=70,c='red',label = 'Pontos da questão')
plt.plot(x,u1,'b-', color='green', label = 'Função U1(x)')
plt.plot(x,u2,'b-',color='orange', label = 'Função U2(x)')
plt.title('Funções obtidas')  # titulo do grafico 
plt.xlabel('X')  #legenda em x
plt.ylabel('Y')  #legenda em y
plt.legend()  # Esse é para apresentar os nomes (label) de cada linha 
plt.grid('True')
plt.show() # para apresentar o grafico
```
