```
import numpy as np
# V^T.a=V^t.y ajuste linear geral 
# X e Y anunciados pela questão
xi = np.array([ -1, -0.8, -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6, 0.8, 1, 1.5, 2.0], dtype='double')
yi = []
for i in range(len(xi)):
    yi.append( xi[i]**4 + 8*xi[i]**2 - 7*xi[i] )
#-------------------------------------------------------------------------------------
# FUNÇÃO DADA PELA QUESTÃO
# f(x)= x**4 + 8*x**2-7*x 
# u(x)= a+ a1*x + a2*x**2 <--- V
V = np.array([ xi**0, xi**1, xi**2], dtype='double').transpose()
##aa=(V^t.V)^1.V^t.y
aa = ((np.linalg.inv((V.transpose()).dot(V))).dot(V.transpose())).dot(yi)
a0 = aa[0]
a1 = aa[1]
a2 = aa[2]
o=[ ' a0', ' a1', ' a2']
print(f'\n\n X={xi}\n\n Y={yi}\n\n V1={V}\n\n a={aa}\n\naa={o}\n\n')
#--------------------------------------------------------------------------------------
# para plotar o grafico
x = np.arange(-1,2,0.000001)
ux = a0 + a1*x + a2*x**2
fx = x**4 + 8*x**2 - 7*x
#>>>>>>>>>>>>>>>>> COMO CRIAR GRAFICOS !!! <<<<<<<<<<<<#
import matplotlib.pyplot as plt #biblioteca de graficos do python
plt.figure(figsize=(12, 12)) #criar o quadro branco de tamanho 10x10
plt.style.use('classic') #o estilo de cores do grafico
plt.scatter(xi,yi,s=70,c='red',label = 'Pontos da questão')
plt.plot(x,ux,'b-', color='green', label = 'Função U(x)')
plt.plot(x,fx,'b-',color='orange', label = 'Função F(x)')
plt.title('Funções Obtidas')  # titulo do grafico 
plt.xlabel('X')  #legenda em x
plt.ylabel('Y')  #legenda em y
plt.legend()  # Esse é para apresentar os nomes (label) de cada linha 
plt.grid('True')
plt.show() # para apresentar o grafico
```
