```
## M√âTODO DE JACOBI ##
#para calcular o numero de condicionamento
import numpy as np
from numpy import linalg as la
#||X|| * ||X^-1|| 
A = np.array([[71,40],[51,30]], dtype='double') #criando a matriz
B = np.linalg.inv(A) # invers√£o da matriz
C = la.norm(A, 1)
D = la.norm(B, 1) #Calculo da norma para 1
print(f'N√∫mero de condicionamento\n Para norma 1:\n{np.multiply(C,D)}\n')  
E = la.norm(A, 2)
F = la.norm(B, 2) #calculo norma 2

print(f'Para norma 2:\n{np.multiply(E,F)}\n')  
G = la.norm(A, np.inf)
H = la.norm(B, np.inf) #calculo norma infinita
print(f'Para norma infinita:\n{np.multiply(G,H)}\n')
# DETERMINANDO AS EQUA√á√ïES
'''
a
71ùë• + 40ùë¶ = 100
51ùë• + 30ùë¶ = 70
'''
def xi(x,y):
    return (100 - 40*y)/71
def yi(x,y):
    return (70 - 51*x)/30
print(f'\nx=(100 - 40*y)/71\n y=(70 - 51*x)/30\n')
# VALORES INICIAIS DO LA√áO DE REPETI√á√ÉO
x0 = 0
y0 = 0
# TOLERANCIA DA QUEST√ÉO
e = 10**(-7)
# CONTADOR PARA O LA√áO
n = 1
l = 10000
# l √© o limite de itera√ß√£o
print(f'x0={x0}\ny0={y0}\nToler√¢ncia={e:.7f}\nLimite de itera√ß√µes={l}\n')
# CONDI√á√ÉO PARA INICIAR O LA√áO
c = True
# FUN√á√ÉO PARA CRIAR TABELA
def ct(c,n,x0,y0,errox,erroy):
    k=' '
    if n==1:
        print(f"Itera√ß√µes{k*12}xi{k*15}yi{k*19}erro(x){k*15}erro(y){k*15}")
        print("-"*100)
    print(f'|{n:3}|         |{x0: 7.7f}|         |{y0: 7.7f}|        |{errox: 7.7f}|        |{erroy: 7.7f}|')
    if c==False and n>=l :
        print("-"*100)
        print("\n\n")
while c :
    # CALCULO DOS VALORES
    x1 = xi(x0,y0)
    y1 = yi(x0,y0)
    # CALCULO DOS ERROS
    errox = abs(x1-x0)/abs(x1)
    erroy = abs(y1-y0)/abs(y1)
    ct(c,n,x0,y0,errox,erroy)
    # INCREMENTA O CONTADOR, E SUBSTITUI OS NOVOS VALORES 
    n += 1
    x0 = x1
    y0 = y1
    if n >= l:
        c=False
        break
    c = errox>e and erroy>e
ct(c,n,x0,y0,errox,erroy)
```
