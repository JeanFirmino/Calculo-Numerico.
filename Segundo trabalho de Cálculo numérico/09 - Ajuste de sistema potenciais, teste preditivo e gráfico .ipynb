```
import numpy as np
# V^T.a=V^t.y ajuste linear geral 
# X e Y anunciados pela questão
xi = np.array([ 10, 15, 20, 25, 30, 35, 40], dtype='double')
yi = np.array([ 19, 47, 65, 103, 132, 194, 295], dtype='double')
#-------------------------------------------------------------------------------------
# FUNÇÃO DADA PELA QUESTÃO
# y1 = bx^a ----> lny = a.lnx + lnb
xn = np.log(xi)
yn = np.log(yi)
V1 = np.array([ xn**1, xn**0], dtype='double').transpose()
##aa=(V^t.V)^1.V^t.y
aa = ((np.linalg.inv((V1.transpose()).dot(V1))).dot(V1.transpose())).dot(yn)
o=[ ' a', ' ln b']
p=[ ' a', ' b']
print(f'''\n\n X={xi}\n\n Y={yi}\n\nDepois de convertidos para lnx e lny,\n\n{xn}\n\n{yn}\n\n V1={V1}\n\n a={aa}\n\naa={o}\n\n
para encontrar o b para e^lnb\n
''')
aa[1]=np.exp(aa[1])
a = aa[0]
b = aa[1]
print(f'\n\naa={aa}\n\naa={p}\n\n')
#--------------------------------------------------------------------------------------
#para a função y=ab^x ------> lny = x.lna +lnb
xv = xi
yv = np.log(yi)
V2 = np.array([ xv**1, xv**0], dtype='double').transpose()
##aa=(V^t.V)^1.V^t.y
ab = ((np.linalg.inv((V2.transpose()).dot(V2))).dot(V2.transpose())).dot(yv)
o=[ ' ln a', ' ln b']
print(f'\n\nX={xv}\n\nY={yv}\n\nV2={V2}\n\nab={ab}\n\nab={o}\n\n')
ab[0]=np.exp(ab[0])
ab[1]=np.exp(ab[1])
a2 = ab[0]
b2 = ab[1]
print(f'\n\nab={ab}\n\nab={p}\n\n')
#---------------------------------------------------------------------------------------
# 9.b
def y1(x,a,b):
    return x*a + b
def y2(x,a2,b2): 
    return x*a2 + b2
print(f'''Para X=27,1 temos;
Y1 = {y1(27.1,a,b)}
Y2 = {y2(27.1,a2,b2)}
Para X=37,8 temos;
Y1 = {y1(37.8,a,b)}
Y2 = {y2(37.8,a2,b2)}
''')
#---------------------------------------------------------------------------------------
# para plotar o grafico
x = np.arange(10,40,0.00001)
y1 = b*x**a
y2 = b2*(a2**x)
#>>>>>>>>>>>>>>>>> COMO CRIAR GRAFICOS !!! <<<<<<<<<<<<#
import matplotlib.pyplot as plt #biblioteca de graficos do python
plt.figure(figsize=(10, 10)) #criar o quadro branco de tamanho 10x10
plt.style.use('classic') #o estilo de cores do grafico
plt.scatter(xi,yi,s=70,c='red',label = 'Pontos da questão')
plt.plot(x,y1,'b-', color='green', label = 'Função Y1(x)')
plt.plot(x,y2,'b-',color='orange', label = 'Função Y2(x)')
plt.title('Funções analisadas')  # titulo do grafico 
plt.xlabel('X')  #legenda em x
plt.ylabel('Y')  #legenda em y
plt.legend()  # Esse é para apresentar os nomes (label) de cada linha 
plt.grid('True')
plt.show() # para apresentar o grafico
```

# Com sklearn

```
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

import numpy as np
# V^T.a=V^t.y ajuste linear geral 
# X e Y anunciados pela questão
xi = np.array([ 10, 15, 20, 25, 30, 35, 40], dtype='double')
yi = np.array([ 19, 47, 65, 103, 132, 194, 295], dtype='double')

```

```
# y = bx^a ==> lny = lnb + a*lnx
X1, y1 = np.log(xi).reshape(-1, 1), np.log(yi).reshape(-1, 1)
regr = linear_model.LinearRegression()
regr.fit(X1, y1)

y_p = regr.predict(X1)

# The coefficients
print('Coefficients: ', regr.coef_)
print('intercept: ', regr.intercept_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y1, y_p))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y1, y_p))
```
