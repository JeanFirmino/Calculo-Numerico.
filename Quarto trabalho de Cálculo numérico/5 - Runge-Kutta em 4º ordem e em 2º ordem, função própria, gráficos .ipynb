```
# Método de Runge-Kutta de ordem 4 # h = 0.5
# Adicionar função para resolver
xk4 = []
yk4 = []
import numpy as np
def fx(x,y):
    return ((3*x*(np.exp(2*x))) - (3*y))
# O método de Runge-Kutta 4º
def rungekutta4(x0,y0,xn,h):
    # Para calcular os passos
    n = round((xn - x0) / h)
    print(f'\n------------------SOLUÇÃO------------------------')
    print(f'----------------------------------------------------')    
    print(f'| x0 |\t | y0 |\t | yn |')
    print(f'----------------------------------------------------')
    for i in range(n+1):
        k1 = h * (fx(x0, y0))
        k2 = h * (fx((x0+h/2), (y0+k1/2)))
        k3 = h * (fx((x0+h/2), (y0+k2/2)))
        k4 = h * (fx((x0+h), (y0+k3)))
        k = (k1+2*k2+2*k3+k4)/6
        yn = y0 + k
        print(f'| {x0} |\t | {y0} |\t | {yn} |' )
        print(f'------------------------------------------------')
        xk4.append(x0)
        yk4.append(y0)
        y0 = yn
        x0 = x0+h
# Entradas 
x0 = 0.50
y0 = 0
# para calcular até o ponto
xn = 3.0
# h passo
h = 0.5
# chamada do método de Euler
rungekutta4(x0,y0,xn,h)
```
