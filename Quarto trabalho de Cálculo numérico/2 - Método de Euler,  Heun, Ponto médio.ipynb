```
# resolver a função 
xne = []
yne = []
import numpy as np
def f(x,y):
    return ((3*x*(np.exp(2*x))) - (3*y))
# Método de Euler
def euler(x0,y0,xn,h):
    n = round((xn - x0)/h)
    print('\n--------------------SOLUÇÃO Euler--------------------')
    print('-------------------------------------------------')    
    print('| i |\t x0 |\t y0 |\t f(x0,y0) |\t yn= y0 + h*f(x0,y0) |')
    print('--------------------------------------------------')
    for i in range(n+1):
        fx = f(x0, y0)
        yn = y0 + ( h*f(x0, y0) )
        print(f'| {i+1} | \t{x0} | \t{y0} | \t{fx} |\t{yn} |')
        print('----------------------------------------------')
        xne.append(x0)
        yne.append(y0)
        y0 = yn
        x0 += h
# Entradas
x0 = 0.50
y0 = 0
# para calcular até o ponto
xn = 3.0
# h passo
h = 0.50
# chamada do método de Euler
euler(x0,y0,xn,h)
```

```
# resolver a função
xnh = []
ynh = []
import numpy as np
def f(x,y):
    return ((3*x*(np.exp(2*x))) - (3*y))
# Método de Heun
def heun(x0,y0,xn,h):
    n = round((xn - x0) / h)
    print('\n--------------------SOLUÇÃO Heun---------------------------------------------------------------------------------------------------------------')
    print('----------------------------------------------------------------------------------------------------------------------------------------------')    
    print('| i | \t\t x0 | \t\t y0 |\t f(x0 ,y0) | \t\t yn = y0 + h * fx | \t\t yn = y0 + ( ( fx + f(xf, y0n) ) / 2 ) * h |')
    print('-----------------------------------------------------------------------------------------------------------------------------------------------')
    for i in range(n+1):
        fx = f(x0, y0)
        y0n = y0 + ( h * fx ) #preditor
        xf = x0 + h
        yn = y0 + ( fx + f(xf, y0n) ) * (h/2) #preditor
        print(f'| {i+1} | \t\t{x0} | \t\t{y0} | \t\t{fx} | \t\t{y0n} \t\t{yn} |')
        print('-------------------------------------------------------------------------------------------------------------------------------------------')
        xnh.append(x0)
        ynh.append(y0)
        y0 = yn
        x0 += h
# Entradas
x0 = 0.50
y0 = 0
# para calcular até o ponto
xn = 3.0
# h passo
h = 0.50
# chamada do método de heun
heun(x0,y0,xn,h)
```

```
# resolver a função
xnm = []
ynm = []
import numpy as np
def f(x,y):
    return ((3*x*(np.exp(2*x))) - (3*y))
# Método do ponto médio
def medio(x0,y0,xn,h):
    # Calculando os passos
    n = round((xn - x0) / h)
    print('\n--------------------SOLUÇÃO Ponto médio--------------------')
    print('-------------------------------------------------')    
    print('| i | x0 | y0 | f(x0,y0) | y(n + 1/2) = y0 + (h/2) * f(x0 ,y0) | x(n + 1/2) = x0 + h/2 | yl = f(x(n + 1/2), y(n + 1/2)) | yn = y0 + ( yl * h ) |')
    print('--------------------------------------------------')
    for i in range(n+1):
        fx = f(x0, y0)
        yf = y0 + (h/2) * fx 
        xf = x0 + h/2
        yl = f(xf, yf)
        yn = y0 + ( yl * h )
        print(f'| {i+1} |\t| {x0} |\t| {y0} |\t| {fx} |\t| {yf} |\t| {xf} |\t| {yl} |\t| {yn} ')
        print('----------------------------------------------')
        xnm.append(x0)
        ynm.append(y0)
        y0 = yn
        x0 += h
# Entradas
x0 = 0.50
y0 = 0
# para calcular até o ponto
xn = 3.0
# h passo
h = 0.50
# chamada do método de ponto medio
medio(x0,y0,xn,h)
```
