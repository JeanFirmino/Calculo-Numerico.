
### h = 0.25



```
# resolver a função 
xne = []
yne = []
import numpy as np
def f(x,y):
    return ((3*x*(np.exp(2*x))) - (3*y))
# Método de Euler
def euler(x0,y0,xn,h):
    n = round((xn - x0) / h)
    print('\n--------------------SOLUÇÃO Euler---------------------------------------------------------------------------------------')
    print('---------------------------------------------------------------------------------------------------------')    
    print('| i | \t | x0 | \t | y0 | \t | f(x0,y0) | \t  | yn= y0 + h*f(x0,y0) |')
    print('----------------------------------------------------------------------------------------------------------------')
    for i in range(n+1):
        fx = f(x0, y0)
        yn = y0 + ( h * fx )
        print(f'| {i+1} | \t | {x0} | \t | {y0} | \t | {fx} | \t | {yn} |')
        print('------------------------------------------------------------------------------------------------------')
        xne.append(x0)
        yne.append(y0)
        y0 = yn
        x0 += h
# Entradas
x0 = 0.50
y0 = 0
# para calcular até o ponto
xn = 3.0
# h passo
h = 0.25
# chamada do método de Euler
euler(x0,y0,xn,h)
```

```
# resolver a função
xnh = []
ynh = []
import numpy as np
def f(x,y):
    return ((3*x*(np.exp(2*x))) - (3*y))
# Método de Heun
def heun(x0,y0,xn,h):
    n = round((xn - x0) / h)
    print('\n--------------------SOLUÇÃO Heun-------------------------------------------------------------------------------------------------')
    print('--------------------------------------------------------------------------------------------------------------------------------------')    
    print('| i | \t | x0 | \t | y0 | \t | f(x0 ,y0) | \t | yn = y0 + h * fx | \t | yn = y0 + ( ( fx + f(xf, y0n) ) / 2 ) * h |')
    print('------------------------------------------------------------------------------------------------------------------------------')
    for i in range(n+1):
        fx = f(x0, y0)
        y0n = y0 + ( h * fx ) #preditor
        xf = x0 + h
        yn = y0 + ( ( fx + f(xf, y0n) ) / 2 ) * h #preditor
        print(f'| {i+1} | \t | {x0} | \t | {y0} | \t | {fx} | \t | {y0n} | \t | {yn} |')
        print('------------------------------------------------------------------------------------------------------------------------')
        xnh.append(x0)
        ynh.append(y0)
        y0 = yn
        x0 += h
# Entradas
x0 = 0.50
y0 = 0
# para calcular até o ponto
xn = 3.0
# h passo
h = 0.25
# chamada do método de heun
heun(x0,y0,xn,h)
```

```
# resolver a função
xnm = []
ynm = []
import numpy as np
def f(x,y):
    return ((3*x*(np.exp(2*x))) - (3*y))
# Método do ponto médio
def medio(x0,y0,xn,h):
    # Calculando os passos
    n = round((xn - x0) / h)
    print('\n--------------------SOLUÇÃO Ponto médio-----------------------------------------------------------------------------------------------------------------------------------')
    print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------')    
    print('| i | \t | x0 | \t | y0 | \t | f(x0,y0) | \t | y(n + 1/2) = y0 + (h/2) * f(x0 ,y0) | \t | x(n + 1/2) = x0 + h/2 | \t | yl = f(x(n + 1/2), y(n + 1/2)) | \t | yn = y0 + ( yl * h ) |')
    print('------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
    for i in range(n+1):
        fx = f(x0, y0)
        yf = y0 + (h/2) * fx 
        xf = x0 + h/2
        yl = f(xf, yf)
        yn = y0 + ( yl * h )
        print(f'| {i+1} | \t | {x0} | \t | {y0} | \t | {fx} | \t | {yf} | \t | {xf} | \t | {yl} | \t | {yn} ')
        print('------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
        xnm.append(x0)
        ynm.append(y0)
        y0 = yn
        x0 += h
# Entradas
x0 = 0.50
y0 = 0
# para calcular até o ponto
xn = 3.0
# h passo
h = 0.25
# chamada do método de ponto medio
medio(x0,y0,xn,h)
```

```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#Função dy/dx
def f(y,x):
    dydx = ((3*x*(np.exp(2*x))) - (3*y))
    return dydx
# Condição inicial
y0 = 0
# intervalo
x = np.linspace(0.5,3.0)
# Para resolver 
y = odeint(f,y0,x)
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 4))
plt.plot( x, y,'b-', linewidth=3, color='green', label = 'Solução Exata')
plt.plot( xne, yne,'o-', color='red', label = 'Euler')
plt.plot( xnh, ynh,'o-', color='orange', label = 'Heun')
plt.plot(xnm,ynm,'o-',color='blue', label = 'Ponto médio')
plt.style.use('classic')
plt.title('Soluções obtidas')
plt.xlabel('X (intervalo)')
plt.ylabel('Y (valores obtidos)')
plt.ylim()
plt.xlim(0.5,3)
plt.legend(loc='upper left')
plt.grid('True')
plt.show()
```
