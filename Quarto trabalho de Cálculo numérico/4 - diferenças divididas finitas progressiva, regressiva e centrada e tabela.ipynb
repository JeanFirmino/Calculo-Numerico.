```
import numpy as np
from sympy import *
from math import*
#x = Symbol('x')
#y = cos( 4*x**3 + x**2 )
#y1 = y.diff( x, n = 1)
#y2 = y1.diff( x, n = 1)
def y1(x):
    return -(12*x**2 + 2*x)*sin(4*x**3 + x**2)
def y2(x):
    return -(24*x + 2)*sin(4*x**3 + x**2) - (12*x**2 + 2*x)**2*cos(4*x**3 + x**2)
# Diferença dividida finita progressiva
def fx(x):
    return np.cos( 4*x**3 + x**2 )
def f11x(x,h):
    return ( fx(x+h) - fx(x) ) / h 
def f21x(x,h):
    return ( fx(x+2*h) - 2*fx(x+h) + fx(x) ) / (h**2)
xi = np.array([0.45,0.65,0.85])
h = 0.20
def criatab(x,fx,f11x,f21x,erro1,erro2):
    if i==0:
        print(f'|X|\t|f(x)|\t\t\t|f1(x)|\t\t\t|f2(x)|\t\t\t|Erro1|\t\t\t|Erro2|')
    print(f'|{x}|\t|{fx}|\t|{f11x}|\t|{f21x}|\t|{erro1}|\t|{erro2}|\t')
# para o as primeiras funções de diferença dividida finita progressiva
for i in range(len(xi)):
    erro1 = abs(y1(xi[i]) - f11x(xi[i],h))/abs(y1(xi[i]))
    erro2 = abs(y2(xi[i]) - f21x(xi[i],h))/abs(y2(xi[i]))
    criatab(xi[i],fx(xi[i]),f11x(xi[i],h),f21x(xi[i],h),erro1,erro2)
print('\n')
# Diferença dividida finita regressiva
def fx(x):
    return np.cos( 4*x**3 + x**2 )
def f11x(x,h):
    return ( fx(x) - fx(x-h) ) / h 
def f21x(x,h):
    return ( fx(x-2*h) - 2*fx(x-h) + fx(x) ) / (h**2)
xi = np.array([0.45,0.65,0.85])
h = 0.20
def criatab(x,fx,f11x,f21x,erro1,erro2):
    if i==0:
        print(f'|X|\t|f(x)|\t\t\t|f1(x)|\t\t\t|f2(x)|\t\t\t|Erro1|\t\t\t|Erro2|')
    print(f'|{x}|\t|{fx}|\t|{f11x}|\t|{f21x}|\t|{erro1}|\t|{erro2}|\t')
# para o as primeiras funções de diferença dividida finita regressiva
for i in range(len(xi)):
    erro1 = abs(y1(xi[i]) - f11x(xi[i],h))/abs(y1(xi[i]))
    erro2 = abs(y2(xi[i]) - f21x(xi[i],h))/abs(y2(xi[i]))
    criatab(xi[i],fx(xi[i]),f11x(xi[i],h),f21x(xi[i],h),erro1,erro2)
print('\n')
# Diferença dividida finita centrada
def fx(x):
    return np.cos( 4*x**3 + x**2 )
def f11x(x,h):
    return ( fx(x+h) - fx(x-h) ) / (2*h)
def f21x(x,h):
    return ( fx(x+h) - 2*fx(x) + fx(x-h) ) / (h**2)
xi = np.array([0.45,0.65,0.85])
h = 0.20
def criatab(x,fx,f11x,f21x,erro1,erro2):
    if i==0:
        print(f'|X|\t|f(x)|\t\t\t|f1(x)|\t\t\t|f2(x)|\t\t\t|Erro1|\t\t\t|Erro2|')
    print(f'|{x}|\t|{fx}|\t|{f11x}|\t|{f21x}|\t|{erro1}|\t|{erro2}|\t')
# para o as primeiras funções de diferença dividida finita centrada
for i in range(len(xi)):
    erro1 = abs(y1(xi[i]) - f11x(xi[i],h))/abs(y1(xi[i]))
    erro2 = abs(y2(xi[i]) - f21x(xi[i],h))/abs(y2(xi[i]))
    criatab(xi[i],fx(xi[i]),f11x(xi[i],h),f21x(xi[i],h),erro1,erro2)
```
