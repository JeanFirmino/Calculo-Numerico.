```
## MÉTODO DE JACOBI ##

# DETERMINANDO AS EQUAÇÕES
def xi(x,y,z):
    return (17-y+2*z)/20

def yi(x,y,z):
    return (-18-3*x+z)/20

def zi(x,y,z):
    return (25-2*x+3*y)/20

# VALORES INICIAIS DO LAÇO DE REPETIÇÃO
x0 = 0
y0 = 0
z0 = 0

# TOLERANCIA DA QUESTÃO
e = 10**(-7)

# CONTADOR PARA O LAÇO

n = 1
l = 10
# l é o limite de iteração

# CONDIÇÃO PARA INICIAR O LAÇO
c = True

# FUNÇÃO PARA CRIAR TABELA
def ct(c,n,x0,y0,z0,errox,erroy,erroz):

    k=' '
    
    if n==1:
    
        print(f"Iterações{k*12}xi{k*15}yi{k*19}zi{k*15}erro(x){k*15}erro(y){k*15}erro(z){k*15}")
    
        print("-"*160)
    
    print(f'|{n:3}|         |{x0: 7.7f}|         |{y0: 7.7f}|          |{z0: 7.7f}|        |{errox: 7.7f}|        |{erroy: 7.7f}|        |{erroz: 7.7f}|')
    
    if c==False:
    
        print("-"*160)
    
        print("\n\n")

while c:
   # CALCULO DOS VALORES
    x1 = xi(x0,y0,z0)
    y1 = yi(x0,y0,z0)
    z1 = zi(x0,y0,z0)
   
# CALCULO DOS ERROS
    errox = abs(x0-x1);
    erroy = abs(y0-y1);
    erroz = abs(z0-z1);

    ct(c,n,x0,y0,z0,errox,erroy,erroz)

    # ICREMENTA O CONTADOR, E SUBSTITUI OS NOVOS VALORES 
    n += 1
    x0 = x1
    y0 = y1
    z0 = z1
    
    if n>l:
        c=False
    c = errox>e and erroy>e and erroz>e

ct(c,n,x0,y0,z0,errox,erroy,erroz)
```
