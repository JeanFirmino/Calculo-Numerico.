### Ponto Fixo

```
# metodo do ponto fixo #
from math import *

'''Importamos a biblioteca math apenas para poder usar algumas fun√ßoes matematicas basicas
como a raiz e o modulo de um valor '''



print(f'A fun√ß√£o √©: ùë•**2 + ùë• - 6\n')
print(f'X = g(X) \n X = 6/x+1\nX**1 = 0\n') #e a tolerancia √© 10**-7'
a = 0  #X inicial
tol = 10**-7 # Tolerancia 
iteracao = 20  #numero de itera√ß√µes 
'''
As condi√ßoes que como x1 inicial, tolerancia e o limite de itera√ß√µes

'''
def f(x):
    return(ùë•**2 + ùë• - 6)

''' Essa fun√ß√£o foi feita apenas para facilitar quando o programa necessitar fazer varios
chamamentos para a mesma fun√ß√£o, podiamos fazer ela diretamente dentro do while, mas 
apenas deixaria mais complicado de trabalhar, pode modificar por quest√£o'''

# reescrever f(x)=0 como x = g(x)
def g(x):
    return(6/(x+1)) #6 + x**2 o sistema bugou  # sqrt(6-x)nessa deu muito certo 6/(x + 1) nessa deu certo

#tabela (VERIFICAR a fun√ß√£o "criartabela() em bisse√ß√£o ")
def ct( cont, xi, a, t, erro):
    if cont==1:
        print(f"Itera√ß√µes{' '*10}xi{' '*16}a{' '*16}f(xi){' '*16}erro{' '*16}")
        print("-"*135)
    print(f'|{cont:3}|        |{xi:.10f}|   |{a:.10f}|      |{t:.10f}|      |{erro:.10f}|')
    if n==False:
        print("-"*135)
        print("\n")


 
'''g(x) = x + a(x)*f(x) tal que a(x)!=0
pelo teorema, ent√£o x = g(x)
'''
cont = 1 # contador para evitar a quantidade de itera√ß√µes sobrepor o limite proposto
conver = True # verificar se √© convergente ou n√£o
n = True # manter o while funcionando 
while n:    
    xi = g(a)  #ponto fixo
    erro = fabs((xi - a) / xi) # calculo do erro relativo, o anterior "a"  menos Xi, o Xi sempre ser√° o novo
    ct( cont, xi, a, f(xi), erro)      #chamamento da fun√ß√£o, logo apos a mudan√ßa de a
    a = xi
    
    cont += 1
        
    if cont > iteracao:   #se a quantidade de itera√ß√µes for maior 
        conver=False
        break
        
    n = abs(f(xi)) > 0 #nao parar se o f(xi) for maior que zero 
ct( cont, xi, a, f(xi), erro) #para fechar a tabela
if conver==True:
    print('A raiz √©: %0.8f \n\n' % xi)
else:
    print('\nN√£o converge\n\n')
```
