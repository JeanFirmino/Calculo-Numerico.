```
from math import*
def f(x):
    return e**(-x**2) - x
def g(x):
    return e**(-x**2) - 2*x
def h(x):
    return (x**5)/10 + x**4 - 3*x**2 + 1
x = []
indicador = []
fx = []
gx = []
hx = []
y = []
v = []
i=-2
while i<=2:
    x.append(i)
    fx.append(f(i))
    gx.append(g(i))
    hx.append(h(i))
    y.append(0)
    v.append(0)
    i+=0.001
i=-2
while i<=2:
    indicador.append(i)
    i+=0.5
#Apresentar o grafico do resultado
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

plt.figure(figsize=(10, 10))
plt.style.use('fivethirtyeight')
line1 = plt.plot(x,fx, '-b', color='blue', label = 'e**(-x**2) - x')
line2 = plt.plot(x,gx, '-b', color='green', label = 'e**(-x**2) - 2*x')
line3 = plt.plot(x,hx, '-b', color='red', label = '(x**5)/10 + x**4 - 3*x**2 + 1')
plt.plot(x,y, '-b', color='black')
plt.plot(v,hx, '-b', color='black')
plt.title('Graficos das Funes')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.xticks(indicador)
plt.show()
```
