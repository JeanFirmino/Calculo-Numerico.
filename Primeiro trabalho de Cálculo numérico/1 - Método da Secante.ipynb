### Método da secante

```
from scipy.optimize import*
def f(x):
    return e**(-x**2) - x
def g(x):
    return e**(-x**2) - 2*x
def h(x):
    return (x**5)/10 + x**4 - 3*x**2 + 1

print(newton(f,0))

print(newton(g,0))

print(newton(h,-15))

print(newton(h,-4))

print(newton(h,-1))

print(newton(h,1))

print(newton(h,3))
#Impressionante como é gigantesco o metodo de newton

print(root_scalar(f,method='secant',x0=0,x1=2))

print(root_scalar(g,method='secant',x0=0,x1=2))

print(root_scalar(h,method='secant',x0=1,x1=3))

print(root_scalar(h,method='secant',x0=1,x1=0))

print(root_scalar(h,method='secant',x0=0,x1=-1))

print(root_scalar(h,method='secant',x0=-1,x1=-3))

print(root_scalar(h,method='secant',x0=-3,x1=-10))
```

```
#>>>> METODO DA SECANTE <<<<#
from math import*

'''Importamos a biblioteca math apenas para poder usar algumas funçoes matematicas basicas
como a raiz e o modulo de um valor '''


# definindo a função 
def f(x):
    return (x**5)/10 + x**4 - 3*x**2 + 1

''' Essa função foi feita apenas para facilitar quando o programa necessitar fazer varios
chamamentos para a mesma função, podiamos fazer ela diretamente dentro do while, mas 
apenas deixaria mais complicado de trabalhar, pode modificar por questão'''


# o metodo da secante
a = -15 #primeiro valor
b = -10 #segundo valor
tol = 0.0000001  #tolerancia 
iter = 20 #iteraçao maxima 
km ='(x**5)/10 + x**4 - 3*x**2 + 1'
print(f'A Função {km}\na={a}\nb={b}\nTolerancia = {tol:.7f}\nLimite de iterações = {iter}\n\n ')

''' condiçoões modificaveis por questão '''



#função para criar a tabela (verificar o metodo da bisseção pra maiores informações sobre a tabela)
def ct(cont, xi,a,t,p,erro):
    k = ' '*20
    if cont==1:
        print(f"Iterações{' '*10}xi{' '*15}f(xi){' '*10}f(a){' '*15}f(b){' '*10}erro{' '*10}")
        print("-"*135)
    print(f'|{cont:3}|           |{xi:.7f}|     |{a:.7f}|     |{t:.7f}|      |{p:.7f}|     |{erro:.7f}|')
    if cont > iter or parada==False:
        print("-"*135)
        print("\n\n")

# metodo da secante 
cont = 1
parada = True
while parada:
    if f(a) == f(b):  #não tem como dividir por zero 
        print('Divisão por zero!')
        break
        
    xi = a - (b-a)*f(a)/( f(b) - f(a) ) #metodo da secante

    erro = fabs((xi - b) / xi)   #calculo do erro relativo 
    
    ct(cont, xi, f(xi),f(a),f(b),erro)
    
    a = b
    b = xi
    cont += 1
        
    if cont > iter:
        print('Não converge')
        break
        
    parada = abs(f(xi)) > tol  ## enquanto xi não for raiz, a repetição prossegue.
ct(cont, xi, f(xi),f(a),f(b),erro)

print(f'\n A Raiz é {xi:0.8f}\n\n')
```
