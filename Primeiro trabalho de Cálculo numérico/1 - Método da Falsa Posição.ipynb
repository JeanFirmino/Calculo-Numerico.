```
# Método da Falsa Posição
# importando a biblioteca matemática
import math as m

# inserindo a função

def f(x):
  return m.sqrt(x) - m.cos(x) - 2

# inserir o intervalo [a,b] 

a = 1

b = 3 

print(f'Insira a função : sqtr(x) - cos(x) -2 ')
print(f'O intervalo [{a},{b}]') 

# para inserir a tolerância 
'''
tol=10**-12
p = '10^-12'
print(f'A Tolerancia é {tol} = {p}')'''
#plotar o grafico de resultados
iteracoes=[]
raiz=[]
# quantidade de iterações e um contador

c = 10
cont = 1
print(f'Números de iterações {c}\n\n\n')

# criar tabela de valores

def criartabela(a,b,xi,erro,t,u,i,cont):

  if cont==1:
    print(f"Iterações{' '*10}a{' '*16}b{' '*16}xi{' '*13}erro{' '*15}f(a){' '*12}f(b){' '*13}f(xi)")
    print("-"*135)
 
  print(f'|{cont:3}|        |{a:.10f}|   |{b:.10f}|   |{xi:.10f}|   |{erro:.10f}|   |{t:.10f}|   |{u:.10f}|   |{i:.10f}|')
 
  if cont==c:
    print("-"*135)
    print("\n\n")

#criar função para o erro

def erro():
  return m.fabs(b-a)/2

# o teorema de bolzano
#while (m.fabs(b-a)/2 > tol) and cont<=c:
if f(a)*f(b)<0:    #Primeiro teste para verificar se podemos usar o metodo da bisseção
  while cont<=c:

    erro = m.fabs(b-a)/2 # cálculo para o erro.

    xi = b - ((f(b)*(b-a))/(f(b)- f(a)))   # Aplicação do método.
    ''''    ### erro relativo ###
    if cont==1:
        xi = b - ((f(b)*(b-a))/(f(b)- f(a)))  
        erro = fabs((xi - xi) / xi)*100
    else:
        xa = xi
        xi = b - ((f(b)*(b-a))/(f(b)- f(a)))  
        erro = fabs((xi - xa) / xi)*100
    '''
    if f(xi)==0:
      break
    else: #caso não tenha raiz faremos aplicamos bolzano novamente
        if f(a)*f(xi)<0:
          criartabela(a,b,xi,erro,f(a),f(b),f(xi),cont)
          b = xi
        elif f(b)*f(xi)<0:
          criartabela(a,b,xi,erro,f(a),f(b),f(xi),cont)
          a = xi
        else:
          print('de acordo com bolzano, não podemos calcular essa raiz')
          break
    iteracoes.append(cont)
    raiz.append(xi)
    cont+=1
else:
  print(f'f(a)*f(b)={f(a)*f(b)}')  #Caso não seja possivel utilizar o método inicialmente.

print(f'o valor da raiz é : {xi} \n\n')

#Apresentar o grafico
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.style.use('fivethirtyeight')
plt.plot(iteracoes,raiz, '-b', color='blue')
plt.scatter(iteracoes, raiz, s= 100, color='red')
plt.title('Quantidade de iterações/Aproximação resultante')
plt.xlabel('Quantidade de iterações')
plt.ylabel('Valores de Xi')
plt.xticks(iteracoes)
#plt.yticks(raiz)
plt.show()
  
print('\n')
#plotar o grafico da função
xis=0
teste = []
resultados = []
while xis<=10:
    teste.append(xis)
    resultados.append(f(xis))
    xis+=1

plt.figure(figsize=(10, 10))
plt.style.use('fivethirtyeight')
plt.plot(teste,resultados, '-', color='blue')
plt.title('Gráfico da Função')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.xticks(teste)
#plt.yticks(resultados)
plt.show()
```
