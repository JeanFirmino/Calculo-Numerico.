```
# resolver a função 
xne = []
yne = []
import numpy as np
def f(x,y):
    return (1+4*x)*((y)**(1/2))
# Método de Euler
def euler(x0,y0,xn,h):
    n = round((xn - x0)/h)
    print('\n--------------------SOLUÇÃO Euler--------------------')
    print('-------------------------------------------------')    
    print('| i | \t\t | x0 | \t\t | y0 |')
    print('--------------------------------------------------')
    for i in range(n+1):
        fx = f(x0, y0)
        yn = y0 + ( h*f(x0, y0) )
        print(f'| {i+1} | \t\t | {x0} | \t\t | {y0} |')
        print('----------------------------------------------')
        xne.append(x0)
        yne.append(y0)
        y0 = yn
        x0 += h
# Entradas
x0 = 0.0
y0 = 1.0
# para calcular até o ponto
xn = 1.0
# h passo
h = 0.20
# chamada do método de Euler
euler(x0,y0,xn,h)
```

```
# resolver a função
xnh = []
ynh = []
import numpy as np
def f(x,y):
    return (1 + 4*x )*(( y )**( 1/2 ))
# Método de Heun
def heun(x0,y0,xn,h):
    n = round((xn - x0) / h)
    print('\n--------------------SOLUÇÃO Heun---------------------------------------------------------------------------------------------------------------')
    print('----------------------------------------------------------------------------------------------------------------------------------------------')    
    print('| i | \t\t | x0 | \t\t | y0 |')
    print('-----------------------------------------------------------------------------------------------------------------------------------------------')
    for i in range(n+1):
        fx = f(x0, y0)
        y0n = y0 + ( h * fx ) #preditor
        xf = x0 + h
        yn = y0 + ( fx + f(xf, y0n) ) * (h/2) #preditor
        print(f'| {i+1} | \t\t | {x0} | \t\t | {y0} |')
        print('-------------------------------------------------------------------------------------------------------------------------------------------')
        xnh.append(x0)
        ynh.append(y0)
        y0 = yn
        x0 += h
# Entradas
x0 = 0.0
y0 = 1.0
# para calcular até o ponto
xn = 1.0
# h passo
h = 0.20
# chamada do método de heun
heun(x0,y0,xn,h)
```

```
# resolver a função
xnm = []
ynm = []
import numpy as np
def f(x,y):
    return (1+4*x)*((y)**(1/2))
# Método do ponto médio
def medio(x0,y0,xn,h):
    # Calculando os passos
    n = round((xn - x0) / h)
    print('\n--------------------SOLUÇÃO Ponto médio--------------------')
    print('-------------------------------------------------')    
    print('| i | \t | x0 | \t | y0 |')
    print('--------------------------------------------------')
    for i in range(n+1):
        fx = f(x0, y0)
        yf = y0 + (h/2) * fx 
        xf = x0 + h/2
        yl = f(xf, yf)
        yn = y0 + ( yl * h )
        print(f'| {i+1} | \t | {x0} | \t | {y0} |')
        print('----------------------------------------------')
        xnm.append(x0)
        ynm.append(y0)
        y0 = yn
        x0 += h
x0 = 0.0
y0 = 1.0
# para calcular até o ponto
xn = 1.0
# h passo
h = 0.20
# chamada do método de ponto medio
medio(x0,y0,xn,h)
```

```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
#Função dy/dx
def f(y,x):
    dydx = (1+4*x)*((y)**(1/2))
    return dydx
# Condição inicial
y0 = 1.0
# intervalo
x = np.linspace(0.0,1.0)
# Para resolver 
y = odeint(f,y0,x)
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 5))
plt.plot( x, y,'b-', linewidth=5, color='green', label = 'Solução Exata')
plt.plot( xne, yne,'o-', color='red', label = 'Euler')
plt.plot( xnh, ynh,'o-', color='orange', label = 'Heun')
plt.plot(xnm,ynm,'o-',color='blue', label = 'Ponto médio')
plt.style.use('classic')
plt.title('Soluções obtidas')
plt.xlabel('X (intervalo)')
plt.ylabel('Y (Valores obtidos)')
plt.ylim()
plt.xlim(0,1)
plt.legend(loc = 'upper left')
plt.grid('True')
plt.show()
```
