```
# Método do Trapézio
# Definir a função de integração
import numpy as np
def f(x):
    return (10*(np.exp(-x))*(np.sin(2*(np.pi)*x)))**2
# Aplicando o método do trapézio
def metodo_do_trapezio( limite_inferior, limite_superior, sub_intervalo):
    # dividindo em intervalos
    h = (limite_superior - limite_inferior) / sub_intervalo
    # encontrando a soma
    integracao = f(limite_inferior) + f(limite_superior)
    for i in range(1,sub_intervalo):
        k = limite_inferior + i*h
        integracao = integracao + 2 * f(k)
    # encontrando o valor da integração final
    integracao = integracao * h/2
    return integracao
for sub_intervalo in range(1,13):
    # Entradas do intervalo e do sub intervalo
    limite_inferior = 0
    limite_superior = 1/2
    # Chamamento do metodo do trapézio
    r1 = metodo_do_trapezio( limite_inferior, limite_superior, sub_intervalo)
    print(f'Integração pelo método do trapézio é igual a {r1}')
```

```
# 1/3 de simpson
# função para integrar
import numpy as np
def f(x):
    return (10*(np.exp(-x))*(np.sin(2*(np.pi)*x)))**2
# implementando a regra de 1/3 de simpson
def regra_de_simpson( limite_inferior, limite_superior, sub_intervalo):
    # Calculando o intervalo
    h = (limite_superior - limite_inferior) / sub_intervalo
    # encontrando a soma 
    integracao = f(limite_inferior) + f(limite_superior)
    for i in range(1,sub_intervalo):
        k = limite_inferior + i*h
        if i%2 == 0:
            integracao = integracao + 2 * f(k)
        else:
            integracao = integracao + 4 * f(k)
    # encontrando valor final da intergração
    integracao = integracao * h/3
    return integracao
for sub_intervalo in range(1,13):
    # entradas da integração
    limite_inferior = 0
    limite_superior = 1/2
    # Chamando a funcao pela regra de 1/3 de simpson
    r2 = regra_de_simpson( limite_inferior, limite_superior, sub_intervalo)
    print(f'A integração pela regra de 1/3 de simpson é {r2}')
```

```
# regra de 3/8 de simpson
# a função de integração
import numpy as np
def f(x):
    return (10*(np.exp(-x))*(np.sin(2*(np.pi)*x)))**2
# O método de 3/8 de simpson
def regra_de_simpson38( limite_inferior, limite_superior, sub_intervalo):
    # calculo do intervalo
    h = (limite_superior - limite_inferior) / sub_intervalo
    # Encontrando a soma
    integracao = f(limite_inferior) + f(limite_superior)
    for i in range(1,sub_intervalo):
        k = limite_inferior + i*h
        if i%2 == 0:
            integracao = integracao + 2 * f(k)
        else:
            integracao = integracao + 3 * f(k)
    # valor final da integral
    integracao = integracao * 3 * h / 8
    return integracao
for sub_intervalo in range(1,13):
    # entradas da integração
    limite_inferior = 0
    limite_superior = 1/2
    # Chamando a funcao pela regra de 3/8 de simpson
    r3 = regra_de_simpson38( limite_inferior, limite_superior, sub_intervalo)
    print(f'A integração pela regra de 3/8 de simpson é {r3}')
```
