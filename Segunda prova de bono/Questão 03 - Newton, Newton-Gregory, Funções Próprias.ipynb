```
# DIFERENÇAS DIVIDIDAS (Newton) para ordem 5 
# para encontrar o produto dos termos
from sympy import *
def produto(i, xn, x):
	prod = 1
	for j in range(i):
		prod = prod * (xn - x[j])
	return prod
# Para carcular a tabela das diferenças divididas
def calculandotab(x, y, n):
	for i in range(1, n):
		for j in range(n - i):
			y[j][i] = ((y[j][i - 1] - y[j + 1][i - 1]) / (x[j] - x[i + j]))
	return y
# aplicando newton, metodo da diferenças divididas
def formula(xn, x, y, n):
	soma = y[0][0]
	for i in range(1, n):
		soma = soma + (produto(i, xn, x) * y[0][i])
	return soma
# apresentando a tabela
def apresentatab(y, n):
	for i in range(n):
		for j in range(n - i):
			print(round(y[i][j], 4), "\t", end = " ")
		print("")
#Dados da questão
#n numero de pares x e y
n = 6
x = [ 12, 14, 16, 18, 20, 22 ]
y = [[0 for i in range(n)] #criando a matriz y de zeros 
        for j in range(n)]
#y[][0] usado para iniciar a tabela das diferenças divididas
y[0][0] = 15.4
y[1][0] = 90.2
y[2][0] = 157.1
y[3][0] = 215.2
y[4][0] = 299.2
y[5][0] = 93.7
# calculando a tabela
y=calculandotab(x, y, n)
# Apresentando a tabela
apresentatab(y, n)
# Valor para interpolar 
v = Symbol('v') # dado na questão
yn5 = formula(v, x, y, n)
print(f'\n Interpolando pelo método de Newton\n Y = {simplify(yn5)}\n')
```
